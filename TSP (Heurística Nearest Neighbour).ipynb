```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rnd=np.random
rnd.seed(0)
n=11
coor_x=list(rnd.rand(n)*100)
coor_y=list(rnd.rand(n)*100)
ciudades=[i for i in range(n)]
arcos=[(i,j) for i in ciudades for j in ciudades if i!=j]
distancias={(i,j):np.hypot(coor_x[i]-coor_x[j],coor_y[i]-coor_y[j]) for i,j in arcos}
```

```python
origen=0
NN=[origen]
while len(NN)<n:
    k=NN[-1]
    nn={(k,j):distancias[k,j] for j in ciudades if k!=j and j not in NN}
    minimo=min(nn[t] for t in nn)
    for i,j in nn:
        if nn[i,j]==minimo:
            NN.append(j)
NN.append(origen)
```

```python
plt.figure(figsize=(12,7))

plt.scatter(coor_x,coor_y, color="red")
for i in ciudades:
    plt.annotate("$%d$"%(i),(coor_x[i],coor_y[i]),(coor_x[i],coor_y[i]+2),color="black")

for i in range(len(NN)-1):
    plt.plot([coor_x[NN[i]],coor_x[NN[i+1]]],[coor_y[NN[i]],coor_y[NN[i+1]]],color="Purple",alpha=5)

plt.xlabel("Distancia x")
plt.ylabel("Distancia y")
plt.title("Ruta Nearest Neighbour")
plt.show()
print(NN)
distancia_total=sum(distancias[NN[i],NN[i+1]] for i in range(len(NN)-1))
print(distancia_total)
```
